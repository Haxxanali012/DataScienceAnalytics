```python
import requests

# Fetch a web page
r = requests.get("https://medium.com/@hassanbalochshar124/heart-diseases-prediction-284c108bf847")
print(r.text)
```

```python
import re

# Remove HTML tags using RegularExpression
pattern = re.compile(r'<.*?>')  # tags look like <...>
print(pattern.sub('', r.text))  # replace them with blank
```

```python
from bs4 import BeautifulSoup
```

```python

# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
#mw-content-container
# Find all articles
summaries = soup.find_all("p", class_="kp kq kr ks b kt ku kv kw kx ky kz la lb lc ld le lf lg lh li lj lk ll lm ln eo bj")
```
